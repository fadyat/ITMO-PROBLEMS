```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.style.use('dark_background')
```

```python
import functools


class ReprWrapper(object):
    def __init__(self, repr_func, func):
        self.repr_func = repr_func
        self.func = func
        functools.update_wrapper(self, func)

    def __call__(self, *args, **kw):
        return self.func(*args, **kw)

    def __repr__(self):
        return self.repr_func(self.func)


def with_repr(repr_func):
    """ decorator for __repr__() for func"""

    def wrap(func):
        return ReprWrapper(repr_func, func)

    return wrap
```

**Sphere function**
$f(x, y) = x^2 + y^2 \qquad
f(0, 0) = 0 \qquad
-\infty \leqslant x, y \leqslant \infty$

```python
@with_repr(lambda x: 'x^2 + y^2')
def f1(v):
    return v[0] ** 2 + v[1] ** 2
```

**Booth's function**
$f(x, y) = (x + 2y - 7)^2 + (2x + y - 5)^2 \qquad
f(1, 3) = 0 \qquad
-10 \leqslant x, y \leqslant 10$

```python
@with_repr(lambda x: '(x + 2y - 7)^2 + (2x + y - 5)^2')
def f2(v):
    return (v[0] + 2 * v[1] - 7) ** 2 + (2 * v[0] + v[1] - 5) ** 2
```

**Matyas function**
$\\$
$f(x, y) = 0.26 (x^2 + y^2) - 0.48 xy \qquad
f(0, 0) = 0 \qquad
-10 \leqslant x, y \leqslant 10$

```python
@with_repr(lambda x: '0.26 * (x^2 + y^2) - 0.48xy')
def f3(v):
    return 0.26 * (v[0] ** 2 + v[1] ** 2) - 0.48 * v[0] * v[1]
```

$\displaystyle \nabla f(\boldsymbol{x}) =
\begin{pmatrix} \displaystyle \frac{\partial f}{\partial x_1}\\ \displaystyle \frac{\partial f}{\partial
x_2}\\ \vdots\\ \displaystyle \frac{\partial f}{\partial x_n} \end{pmatrix}$

$\displaystyle {\frac  {\partial f}{\partial x_{k}}}(x_{1},\cdots ,x_{n})=\lim _{{\Delta x\to 0}}{\frac  {f(x_{1},\ldots ,x_{k}+\Delta x,\ldots ,x_{n})-f(x_{1},\ldots ,x_{k},\ldots ,x_{n})}{\Delta x}}.$

```python
delta_x = 1e-6


def derivative(f, v, i):
    upd_v = np.copy(v)
    upd_v[i] += delta_x
    return (f(upd_v) - f(v)) / delta_x


def gradient(f, v):
    return np.array([derivative(f, v, i) for i in range(len(v))])
```

```python
def with_picture(need_picture=None):
    """ decorator to render functions """

    def actual_decorator(func):
        @functools.wraps(func)
        @mpl.rc_context({
            'figure.figsize': (24, 16),
            'figure.facecolor': 'white',
            'figure.dpi': 80,
            'lines.linestyle': '--',
            'legend.fontsize': 0
        })
        def draw(*args, **kwargs):
            if not need_picture:
                return func(*args, **kwargs)

            xk = func(*args, **kwargs)
            function = kwargs.get("f") if kwargs.get("f") else args[0]
            xk_x = xk[:, 0]
            xk_y = xk[:, 1]
            xk_z = function(xk.T)

            lim = max(max(xk_x), max(xk_y)) + 2
            x = np.linspace(-lim, lim, 300)
            y = np.linspace(-lim, lim, 300)
            x, y = np.meshgrid(x, y)
            z = function(np.array([x, y]))

            axis_lim = (-lim, 2.5 * lim)
            axis = np.linspace(axis_lim[0], axis_lim[1], 2)
            zeros = np.zeros(2)
            figure = plt.figure(constrained_layout=True)

            norm_view = figure.add_subplot(121, projection='3d')
            norm_view.plot_surface(x, y, z, alpha=0.7, cmap='cool')
            norm_view.plot(xk_x, xk_y, xk_z, color='white', marker='.')
            norm_view.plot(axis, zeros, zeros, color='grey')
            norm_view.plot(zeros, axis, zeros, color='grey')
            norm_view.scatter([0, axis_lim[1]], [axis_lim[1], 0], [0, 0], color='grey', marker='>', s=45)
            norm_view.axis('off')

            upper_view = figure.add_subplot(122, aspect='equal')
            upper_view.contourf(x, y, z, alpha=.7, cmap='cool')
            upper_view.plot(xk_x, xk_y, marker='.', color='white')
            return xk

        return draw

    return actual_decorator
```

```python
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.4f' % x)
from IPython.display import display
from IPython.display import HTML


class Repository(object):
    """ Repository for saving gradient descent iterations """
    methods = {}
    id = 0

    @classmethod
    def create(cls, method_name, immutable_data, mutable_data, save_data):
        """ Creating data for a new method """
        if not save_data:
            return

        method_name = f'{method_name}_{cls.id}'
        cls.methods[method_name] = {
            'iters': mutable_data,
            'main': immutable_data
        }
        cls.id += 1
        return method_name

    @classmethod
    def save(cls, method_name, args, save_data):
        """ Saving method iteration """
        if not save_data:
            return

        iters = cls.methods[method_name]['iters']
        for i, arg in enumerate(iters.keys()):
            iters[arg].append(args[i])

    @classmethod
    def visualize(cls, display_iters):
        """ Visualize all algorithms results """
        for method_name, values in cls.methods.items():
            extra_info = {
                'method_name': method_name,
                'iterations': len(values['iters']['xk']) - 1,
                'x0': [values['iters']['xk'][0]],
                'xn': [values['iters']['xk'][-1]]
            }
            extra_info.update(values['main'])
            main = pd.DataFrame(extra_info)
            display(HTML(main.to_html(index=False)))
            if display_iters:
                display(pd.DataFrame(values['iters']))

    @classmethod
    def clear(cls):
        """ Clearing all data """
        cls.methods.clear()
```

## 1. Gradient descent with constant learning rate

$x_{k+1} = x_k - \lambda \nabla f(x_k), \qquad \lambda = const$

```python
@with_picture(True)
def gradient_descent_const(f, x0, learning_rate, tolerance=1e-4, n_iter=1000,
                           save_data=False):
    xk = np.array([x0], dtype=float)
    method_name = Repository.create('const',
                                    immutable_data={
                                        'function': repr(f),
                                        'tolerance': tolerance,
                                        'learning_rate': learning_rate
                                    },
                                    mutable_data={
                                        'xk': [x0],
                                        'fk': [f(x0)]
                                    },
                                    save_data=save_data)

    for _ in range(n_iter):
        step = learning_rate * gradient(f, xk[-1])
        if np.all(np.abs(step) < tolerance):
            break

        xk1 = xk[-1] - step
        xk = np.append(xk, np.array([xk1]), axis=0)
        Repository.save(method_name, [xk1, f(xk1)], save_data)

    return xk
```

## 2. Gradient descent with "step crushing"

$\displaystyle f(x_{k+1}) = f(x_k - \lambda_{k}\nabla f(x_k)) \leqslant f(x_k) - \varepsilon \lambda_{k} \| \nabla f(x_k) \|^2, \ \ \displaystyle \varepsilon \in (0, 1)$

$False$: $\lambda_k \rightarrow \frac{\lambda_k}{\delta}, \ \ \delta \in (0, 1)$

```python
from numpy.linalg import norm


@with_picture(True)
def gradient_descent_step_crushing(f, x0, learning_rate=1, tolerance=1e-4, n_iter=1000, eps=0.1, delta=0.95,
                                   save_data=False):
    xk = np.array([x0], dtype=float)
    fxk = f(x0)
    i, lr = 0, learning_rate
    method_name = Repository.create('step_crushing',
                                    immutable_data={
                                        'function': repr(f),
                                        'tolerance': tolerance,
                                        'eps': eps,
                                        'delta': delta
                                    },
                                    mutable_data={
                                        'xk': [x0],
                                        'fk': [fxk],
                                        'learning_rates': [learning_rate]
                                    },
                                    save_data=save_data)

    while i < n_iter:
        step = lr * gradient(f, xk[-1])
        if np.all(np.abs(step) < tolerance):
            break

        xk1 = xk[-1] - step
        fxk1 = f(xk1)
        if fxk1 - fxk > -eps * lr * (norm(gradient(f, xk[-1]), 2) ** 2):
            lr *= delta
            continue

        xk = np.append(xk, np.array([xk1]), axis=0)
        fxk = fxk1
        Repository.save(method_name, [xk1, fxk1, lr], save_data)
        lr = learning_rate
        i += 1

    return xk
```

```python
dummy = gradient_descent_step_crushing(f2, np.array([10, 10], dtype=float), 1, save_data=True)
```

## 3. Gradient descent with 1D optimization techniques

$\displaystyle \lambda_k = \arg_{\lambda \in [0, +\inf)} \min \ f(x_k - \lambda \nabla f(x_k))$

```python
from opt_algo.opt_1d import *


@with_picture(True)
def gradient_descent_opt1d(f, opt_method, x0, tolerance=1e-4, n_iter=1000,
                           save_data=False):
    xk = np.array([x0], dtype=float)
    l, r = 0, 1e6

    method_name = Repository.create('opt1d',
                                    immutable_data={
                                        'function': repr(f),
                                        'opt_method': str(opt_method).split()[1],
                                        'tolerance': tolerance
                                    },
                                    mutable_data={
                                        'xk': [x0],
                                        'fk': [f(x0)],
                                        'learning_rates': [0]
                                    },
                                    save_data=save_data)

    for _ in range(n_iter):
        learning_rate = opt_method(lambda lr: f(xk[-1] - lr * gradient(f, xk[-1])), l, r, eps=tolerance)
        step = learning_rate * gradient(f, xk[-1])
        if np.all(np.abs(step) < tolerance):
            break

        xk1 = xk[-1] - step
        xk = np.append(xk, np.array([xk1]), axis=0)
        Repository.save(method_name, [xk1, f(xk1), learning_rate], save_data)
    return xk
```

## 4. Fletcher-Reeves Method

${\displaystyle
\vec{x}_{k+1} = \vec{x}_{k} - \lambda_{k+1} \vec{S}_{k} \qquad
\vec{S}_{k+1}=\nabla f({\vec{x}_{k+1}})+\omega _{k+1}{\vec {S}}_{k} \qquad
\omega _{k+1}={\frac {\|\nabla f({\vec{x}_{k+1}})\|^{2}}{\|\nabla f({\vec {x}}_{k})\|^{2}}}}\qquad
\lambda_k = \arg_{\lambda \in [0, +\inf)} \min \ f(x_k - \lambda S_k)
$

```python
@with_picture(True)
def gradient_descent_fletcher_reeves(f, opt_method, x0, tolerance=1e-4, n_iter=1000,
                                     save_data=False):
    xk = np.array([x0], dtype=float)
    wk, sk = 0, gradient(f, x0)
    l, r = 0, 1e6

    method_name = Repository.create('fletcher_reeves',
                                    immutable_data={
                                        'function': repr(f),
                                        'opt_method': str(opt_method).split()[1],
                                        'tolerance': tolerance
                                    },
                                    mutable_data={
                                        'xk': [x0],
                                        'fk': [f(x0)],
                                        'sk': [sk],
                                        'wk': [wk],
                                        'learning_rates': [0]
                                    },
                                    save_data=save_data)

    for _ in range(n_iter):
        learning_rate = opt_method(lambda lr: f(xk[-1] - lr * sk), l, r, eps=tolerance)
        step = learning_rate * sk
        if np.all(np.abs(step) < tolerance):
            break

        xk1 = xk[-1] - step
        wk1 = (norm(gradient(f, xk1), 2) ** 2) / (norm(gradient(f, xk[-1]), 2) ** 2)
        sk1 = gradient(f, xk1) + wk1 * sk
        sk, wk = sk1, wk1
        xk = np.append(xk, np.array([xk1]), axis=0)
        Repository.save(method_name, [xk1, f(xk1), sk1, wk1, learning_rate], save_data)
    return xk
```
